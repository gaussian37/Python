# read xml data

Let's practice with weather forcast data of South Korea.

```python
from bs4 import BeautifulSoup
import urllib.request as req
import os.path
```

Download the weather forecast information. If already downloaded, skip it.

South Korea weather forcast rss address : http://www.weather.go.kr/weather/lifenindustry/sevice_rss.jsp

```python
url = "http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnID=108"
savename = "./data/forecast.xml"

if not os.path.exists(savename):
    req.urlretrieve(url, savename)
    print('Check download')
```

print the xml result with applying indent.

```python
from xml.etree import ElementTree

def indent(elem, level=0):
    i = "\n" + level*"  "
    j = "\n" + (level-1)*"  "
    if len(elem):
        if not elem.text or not elem.text.strip():
            elem.text = i + "  "
        if not elem.tail or not elem.tail.strip():
            elem.tail = i
        for subelem in elem:
            indent(subelem, level+1)
        if not elem.tail or not elem.tail.strip():
            elem.tail = j
    else:
        if level and (not elem.tail or not elem.tail.strip()):
            elem.tail = j
    return elem        

root = ElementTree.parse(savename).getroot()
indent(root)
ElementTree.dump(root)
```

read the xml file with Beautifulsoup

```python
xml = open(savename, 'r', encoding="utf-8")
soup = BeautifulSoup(xml, "html.parser")
```

Check the location

```python
info = {}
for location in soup.find_all("location"):
    loc = location.find("city").string    
    weather = location.find_all("tmn")
    
    # check the duplication
    if not (loc in info):
        info[loc] = []

    for tmn in weather:
        info[loc].append(tmn.string)
```

```python
print(info)
```

Write the result as text file

```python
with open("./data/forecast.txt", "wt") as f:
    for loc in sorted(info.keys()):
        print('+', loc)
        f.write(str(loc) + '\n')
        for n in info[loc]:
            print('-', n,)
            f.write('\t' + str(n) + '\n')
    
```
