# Matplotlib

- pyplot 객체를 사용하여 데이터를 표시
- pyplot 객체에 그래프들을 쌓은 다음 plt.show()를 이용하여 출력

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  

X = np.arange(100)
Y = X**2
plt.plot(X, Y)
plt.show()
```

- Graph는 기본 Figure 객체에 생성됨 (따로 설정하지 않으면 기본 Figure에 그래프 누적됨)
- pyplot 객체 사용 시, 기본 Figure에 그래프가 그려짐

```python
x1 = np.arange(100)
y1 = np.sin(x1)

x2 = np.arange(100)
y2 = np.cos(x2)

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.show()
```

## Figure 지정

- Figure 지정하여 그래프 그리기
- Figure ⊃ Axes ⊃ Graph

```python
fig = plt.figure()
fig.set_size_inches(10, 5) # 크기 지정
ax1 = fig.add_subplot(1, 2, 1) # 행 갯수, 열 갯수, 몇 번째
ax2 = fig.add_subplot(1, 2, 2)

ax1.plot(x1, y1, c = "b")
ax2.plot(x2, y2, c = "g")
plt.show()
```

## LineStyle 지정

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  

x = np.arange(100)
y1 = 0.5 * x
y2 = 2 * x
plt.plot(x, y1, lineStyle = "dashed")
plt.plot(x, y2, lineStyle = "dotted")
```

## Title 지정

- Pyplot에 title 함수 사용, figure의 subplot 별 입력 가능

```python
plt.plot(x, y1, color = "b", linestyle = "dashed")
plt.plot(x, y2, color = "r", linestyle = "dotted")
plt.title("Two lines")
plt.xlabel("X value")
plt.ylabel("Y value")
plt.show()
```

- Latex 타입의 표현도 가능 (수식 표현 가능)

```python
plt.title("$y = \\frac{a}{b}x + c$")
plt.show()
```

## legend(도표 범례) 작성하기

- legend 함수로 범례를 표시함, loc 위치 등 속성 지정

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  

x = np.arange(100)
y1 = 0.5 * x
y2 = 2 * x

plt.plot(x, y1, label = "$y = \\frac{1}{2}x$")
plt.plot(x, y2, label = "$y = 2x$")
plt.legend(shadow = True, fancybox = True, loc = "upper right")
```

## 기타 그래프 꾸미기

- Graph에 텍스트 및 기호 넣기

```python
plt.plot(x, y1, label = "$y = \\frac{1}{2}x$")
plt.plot(x, y2, label = "$y = 2x$")
plt.legend(shadow = True, fancybox = True, loc = "upper right")
plt.text(80, 30, "Line1")
plt.annotate("Line2", xy=(60, 125), xytext=(40, 130), 
             arrowprops = dict(facecolor = "black", shrink = 0.05))
plt.show()
```

- Graph에 격자 넣기
- xy축 범위 조정하기

```python
plt.plot(x, y1, label = "$y = \\frac{1}{2}x$")
plt.plot(x, y2, label = "$y = 2x$")
plt.legend(shadow = True, fancybox = True, loc = "upper right")
plt.text(80, 30, "Line1")
plt.annotate("Line2", xy=(60, 125), xytext=(40, 130), 
             arrowprops = dict(facecolor = "black", shrink = 0.05))

# Graph에 격자 넣기
plt.grid(linewidth = 0.5, linestyle = "--")
# xy축 범위 조정하기
plt.xlim(-50, 120)
plt.ylim(-50, 250)

plt.show()
```

# Matplotlib with Pandas

Boston House Price 예제를 통하여 집값 예측을 해보겠습니다.

X변수 13개
- CRIM : 도시 별 1인당 범죄율
- ZN : 25,000 평방 피트를 초과하는 거주 지역의 비율
- INDUS : 비소매 상업지역이 점유하고 있는 토지의 비율
- CHAS : 찰스강에 대한 변수 (강의 경계에 위치한 경우는 1, 아니면 0)
- NOX : 10ppm 당 농축 일산화 질소
- RM : 주택 1가구당 평균 방의 갯수
- AGE : 1940년 이전에 건축된 소유 주택의 비율
- DIS : 5개의 보스턴 직업 센터까지의 접근성 지수
- RAD : 방사형 도로까지의 접근성 지수
- TAX : 10,000 달러 당 재산세율
- PTRATIO : 도시 별 학생/교사 비율
- B : 도시 별 흑인의 비율
- LSAT : 모집단의 하위계층의 비율

Y변수 1개
- MEDV : 본인 소유의 주택가격(중앙값), (단위 : $1,000)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_url = 'data/housing.data' #Data URL
df_data = pd.read_csv(data_url, sep='\s+', header = None) #csv 타입 데이터 로드, separate는 빈공간으로 지정하고, Column은 없음
df_data.columns = [
    'CRIM','ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO' ,'B', 'LSTAT', 'MEDV'] 
# Column Header 이름 지정
df_data.head()
```

```python
fig = plt.figure()
ax = []
for i in range(1,5):
    ax.append(fig.add_subplot(2,2,i))
    
ax[0].scatter(df_data["CRIM"], df_data["MEDV"])
ax[1].scatter(df_data["PTRATIO"], df_data["MEDV"])
ax[2].scatter(df_data["AGE"], df_data["MEDV"])
ax[3].scatter(df_data["NOX"], df_data["MEDV"])
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10.0, 10.0, forward=True)
ax = []
for i in range(1,5):
    ax.append(fig.add_subplot(2,2,i))

columns = ["CRIM", "PTRATIO", "AGE", "NOX"]
colors = ["b", "g", "c", "r"]
for i in range(4):
    ax[i].scatter(df_data[columns[i]], df_data["MEDV"], color=colors[i], label=columns[i])
    ax[i].legend()
    ax[i].set_title(columns[i])

plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

```python
df_data.plot()
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(5,5)
ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)
ax_1.plot(df_data["MEDV"])
ax_2.hist(df_data["MEDV"], bins=50)
ax_1.set_title("House price MEDV")
ax_2.set_title("House price MEDV")

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
scale_data = std_scaler.fit_transform(df_data)
scale_data
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.boxplot(scale_data, labels=df_data.columns)
fig.set_size_inches(10, 10)
plt.show()
```
